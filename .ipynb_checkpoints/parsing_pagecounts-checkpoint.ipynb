# Parsing pagecounts-cleaned data for traffic

```python
import pandas as pd
from glob import glob
import pickle
import subprocess
import collections
from statistics import median
```

```python
# loading data (cohort and their neighbors)
cohort_dict = pickle.load(open('data/cohorts/cohort_dict(parsing_pagecounts.ipynb).p', 'rb'))
hoax_neighbors = pickle.load(open('data/topics(neighbors)/hoax_neighbors.p', 'rb'))
nonhoax_neighbors = pickle.load(open('data/topics(neighbors)/cohort_neighbors.p', 'rb'))
```

```python
def parse_pagecounts():
    for key, value in cohort_dict.items():
        print('*'*100)
        print('creation date: ', value['creation_date'])
        year = str(value['creation_date'])[0:4]
        month = str(value['creation_date'])[4:6]
        day = str(value['creation_date'])[6:8]
        
        #creating sets for neighbors for faster search times
        hoax_outneighbors_set = set(hoax_neighbors[str(value['hoax_id'])]['outlinks'])
        
        nonhoax_outneighbors_set = set()
        for n in value['nonhoax_id']:
            nonhoax_outneighbors_set.update(nonhoax_neighbors[str(n)]['outlinks'])
        
        files = sorted(glob('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts*.gz'
                            .format(Y = year,M = month)))
        
        for x in files:
            if('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts-{Y}{M}{D}'
               .format(Y = year,M = month, D = day) in x):
                CDIdx_first = files.index(x) # index of the first occurance of the creation date files
                break
                
        for x in reversed(files):
            if('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts-{Y}{M}{D}'
               .format(Y = year,M = month, D = day) in x):
                CDIdx_last = files.index(x) # index of the last occurance
                break
                
        hoax_views_before = collections.defaultdict(int)
        nonhoax_views_before = collections.defaultdict(int)
        
        # iterating through compressed files starting from creation date going backwards
        temp_idx = CDIdx_first
        finished = False
        count = 0
        cur_day = day
        
        while(finished == False):
            if(temp_idx != 0): # if file is not at the start of the month (don't have to iterate another folder)
                temp_idx -= 1
                file_path = files[temp_idx]
                
                if(cur_day != file_path[82:84]): # in case of day change
                    count+=1
                    if count == 8:   # value should be 8
                        break
                    else:
                        cur_day = file_path[82:84]
                        print('collecting page view data for: ', file_path[76:84])
                   
                print(file_path)
                for line in subprocess.Popen(['zcat'], 
                              stdin = open(file_path), 
                              stdout = subprocess.PIPE).stdout:
                    
                    split_line = line.split()
#                     try:
                    if len(split_line) == 2:
                        if split_line[0] in hoax_outneighbors_set:
                            hoax_views_before[split_line[0]] += int(split_line[1])
                        if split_line[0] in nonhoax_outneighbors_set:
                            nonhoax_views_before[split_line[0]] += int(split_line[1])
#                     except IndexError:
#                         pass
                    
            else: # if file is at the start of the month
                if month != '01':
                    month_prior = str(int(month)-1).zfill(2)
                    files = sorted(glob('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts*.gz'
                                        .format(Y = year,M = month_prior)))
                    if not files:   # sometimes the month after is empty and a loop keeps going
                        break       # .... in case the month is empty that means it's not needed (7 days looped)
                    if count == 7:
                        break
                else:
                    month_prior = str(12)
                    year_prior = str(int(year)-1)
                    files = sorted(glob('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts*.gz'
                                        .format(Y = year_prior,M = month_prior)))
                    if not files:
                        break
                        
                temp_idx = len(files) # iterate from the end of the month going backwards
         
        
        hoax_views_after = collections.defaultdict(int)
        nonhoax_views_after = collections.defaultdict(int)
        
        # iterating through compressed files starting from creation date going forwards
        files = sorted(glob('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts*.gz'
                            .format(Y = year,M = month)))
        temp_idx = CDIdx_last
        count = 0
        cur_day = day
        while(True):
            if(temp_idx != len(files)-1): # if file is not at the end of the month (don't have to iterate another folder)
                temp_idx += 1
                file_path = files[temp_idx]
                
                if(cur_day != file_path[82:84]): # if there's a day change
                    count+=1
                    if count == 8:   # value should be 8
                        break
                    else:
                        cur_day = file_path[82:84]
                        print('collecting page view data for: ', file_path[76:84])
                        
                        
                print(file_path)        
                for line in subprocess.Popen(['zcat'], 
                              stdin = open(file_path), 
                              stdout = subprocess.PIPE).stdout:
                    
                    split_line = line.split()
#                     try:
                    if len(split_line) == 2:
                        if split_line[0] in hoax_outneighbors_set:
                            hoax_views_after[split_line[0]] += int(split_line[1])
                        if split_line[0] in nonhoax_outneighbors_set:
                            nonhoax_views_after[split_line[0]] += int(split_line[1])
#                     except IndexError:
#                         pass
                    
            else: # if file is at the end of the month
                if month != '12':
                    month_after = str(int(month)+1).zfill(2)
                    files = sorted(glob('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts*.gz'
                                        .format(Y = year,M = month_after)))
                    if not files:   # sometimes the month after is empty and a loop keeps going
                        break       # .... in case the month is empty that means it's not needed (7 days looped)
                    if count == 7:
                        break
                else:
                    month_after = str(1).zfill(2)
                    year_after = str(int(year)+1)
                    files = sorted(glob('/data/personal/anis/pagecounts-raw/{Y}/{Y}-{M}/cleaned-filtered-pagecounts*.gz'
                                        .format(Y = year_after,M = month_after)))
                    if not files:
                        break
                    
                temp_idx = -1 # iterate from the start of the month going forwards
    
        cohort_dict[key]['hoax_views_before'] = hoax_views_before
        cohort_dict[key]['hoax_views_after'] = hoax_views_after
        cohort_dict[key]['nonhoax_views_before'] = nonhoax_views_before
        cohort_dict[key]['nonhoax_views_after'] = nonhoax_views_after
                
parse_pagecounts()
```

```python
# calculating relative volumes (this time getting delta v/v for each nonhoax in cohort)
for key, value in cohort_dict.items():
#-----------------------------------------------------------------------------------
# for hoaxes
    print("Calculating relative volume change for:", value['hoax_title'],"..." )
    if len(value['hoax_views_before']) == 0:
        v_b_hoax = 0
    else:
        v_b_hoax = median(list(value['hoax_views_before'].values()))
        
    if len(value['hoax_views_after']) == 0:
        v_a_hoax = 0        
    else:
        v_a_hoax = median(list(value['hoax_views_after'].values()))
        
    if (v_a_hoax + v_b_hoax) != 0:
        rel_vol_hoax = (v_b_hoax - v_a_hoax) / (v_a_hoax + v_b_hoax)
    else:
        rel_vol_hoax = 0
    
    cohort_dict[key]['hoax_relative_vol'] = rel_vol_hoax  
#-----------------------------------------------------------------------------------
# for nonhoaxes
    print("Calculating relative volume change for the cohort ...")
    nonhoax_before = []
    nonhoax_after = []
    cohort_dict[key]['nonhoax_relative_vol'] = []
    
    for nonhoax in value['nonhoax_id']:
        # fetching for those neighbors values from the cohort dict
        for outneighbor in nonhoax_neighbors[str(nonhoax)]['outlinks']:
            nonhoax_before.append(value['nonhoax_views_before'][outneighbor])
            nonhoax_after.append(value['nonhoax_views_after'][outneighbor])
            
        v_b_nonhoax = median(nonhoax_before)
        v_a_nonhoax = median(nonhoax_after)
        
        if (v_a_nonhoax + v_b_nonhoax) != 0:
            rel_vol_nonhoax = (v_b_nonhoax - v_a_nonhoax) / (v_a_nonhoax + v_b_nonhoax) # positive indicates precedes
        else:
            rel_vol_nonhoax = 0
            
        cohort_dict[key]['nonhoax_relative_vol'].append(rel_vol_nonhoax)
        
    print("*"*50)
#-----------------------------------------------------------------------------------
```

```python
pickle.dump(cohort_dict, open('data/relative_volumes/relative_volumes.p', 'wb'))
```
